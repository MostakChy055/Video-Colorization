```python
root_dir = '/mnt/hyzhao/Documents/datasets/color_benchmark/image/ILSVRC2012_img'
save_dir = os.path.join(root_dir, 'cval1k')
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
    
for i in range(len(filelist)):
    srcp = os.path.join(os.path.join(root_dir, 'val'), filelist[i].split(' ')[0][1:])
    dstp = os.path.join(save_dir, filelist[i].split(' ')[0][1:].split('/')[1])
    shutil.copy(srcp, dstp)
    print(filelist[i].split(' ')[0][2:])
```
