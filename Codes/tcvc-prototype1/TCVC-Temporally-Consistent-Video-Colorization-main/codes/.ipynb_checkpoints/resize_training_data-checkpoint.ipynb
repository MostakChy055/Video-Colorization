```python
import os
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mkdirss(path):
    if os.path.isdir(path) is False:
        os.makedirs(path)
```

```python
list_filename = 'lists/videvo_train.txt'
with open(list_filename) as f:
    video_list = [line.rstrip() for line in f.readlines()]
```

```python
input_path = '/mnt/hyzhao/Documents/datasets/color_benchmark/video/videvo'
output_path = '/mnt/hyzhao/Documents/datasets/color_benchmark/video/videvo_resized300'
```

```python
H, W = (300,300)
for i in range(len(video_list)):
    video = video_list[i]
    video_path = os.path.join(input_path, video)
    img_list = glob.glob(os.path.join(video_path, '*.jpg'))
    for j in range(len(img_list)):
        im = cv2.imread(img_list[j])
        im_resized = cv2.resize(im, (W, H))
        im_name = os.path.basename(img_list[j]).split('.')[0]
        new_name = im_name + '.png'
        new_path = os.path.join(output_path, video, new_name)
        mkdirss(os.path.join(output_path, video))
        print(new_path)
#         break
#     break
```
