```python
import glob
import os
import numpy as np
import pickle
```

```python
list_filename = 'lists/DAVIS_test.txt'
with open(list_filename) as f:
    video_list = [line.rstrip() for line in f.readlines()]
```

```python
video_list
```

```python
len(video_list)
```

```python
np.random.choice(video_list, 4)
```

```python
oo = {}
oo['keys'] = None
```

```python
oo
```

```python
key_v = []
for i in range(len(video_list)):
    video = video_list[i]
    img_list = sorted(os.listdir(os.path.join('/mnt/hyzhao/Documents/datasets/color_benchmark/video/DAVIS-2017-trainval-480p/DAVIS/JPEGImages/480p', video)))
    for j in range(len(img_list)):
        key = video + '_' + img_list[j].split('.')[0] + '_' + str(len(img_list)-1)
        key_v.append(key)
#         print(key)
#     break
print(len(key_v))
```

```python
oo['keys'] = key_v
```

```python
fw = open('lists/DAVIS_trainval_keys.pkl', 'wb')
pickle.dump(oo, fw)
```
